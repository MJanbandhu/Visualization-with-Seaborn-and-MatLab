```python
# import basic libarraies
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
# seaborn has a inbuilt dataset
sns.get_dataset_names()
```

```python
# load iris data
data=sns.load_dataset("iris")
```

```python
data
```

# Relational plots
Relational plots are used to understand relationship between two variables
* line plot
* scatter plot
* relplot

```python
# line plot
sns.lineplot(data=data,x='sepal_length',y='sepal_width',color='pink',ci=None)
```

```python
data.species.unique()
```

```python
sns.lineplot(data=data,x='sepal_length',y='sepal_width',ci=None,hue='species')
```

# scatter plot

```python
data.to_csv("iris")
```

```python
sns.scatterplot(data=data,x='sepal_length',y='sepal_width')
plt.show()
```

```python
sns.scatterplot(data=data,x='sepal_length',y='sepal_width',hue='species',palette='rainbow')
plt.show()
```

# rel plot

```python
# relplot by default it is scatter plot
# relplot can be converted into line plot by giving kind='line'
sns.relplot(data=data,x='petal_length',y='sepal_width')
```

```python
sns.relplot(data=data,x='petal_length',y='sepal_width',kind='line',ci=None)
```

# categorical plots
It is used to analyse one numerical and one categorical column.
* count plot
* box plot
* barplot
* violin plot
* strip plot
* swarm plot
* cat plot

```python
# count plot
# It just gives count of categories
sns.countplot(data=data,x='species')
```

```python
# bar plot
sns.barplot(data=data,x='species',y='petal_width')
plt.show()
```

```python
# Box plot
# box plot ois used to identify outliers
sns.boxplot(data=data,x='sepal_length')
```

```python
sns.boxplot(data=data,x='species',y='sepal_length')
```

```python
# violin plot
sns.violinplot(data=data,x='sepal_width')
```

```python
sns.violinplot(data=data,x='species',y='sepal_width')
```

```python
# strip plot
sns.stripplot(data=data,x='species',y='sepal_width',hue='species')
```

```python
sns.swarmplot(data=data,x='species',y='sepal_width',hue='species')
```

```python
# cat plot
#using cat plot we can plot box,violin,strip,swarm
sns.catplot(data=data,x='species',y='sepal_length',kind='swarm')
```

# Distribution plots

```python
# histplot
# dist plot
# thse two graphs are used to check distribution of data
```

```python
sns.histplot(data=data,x='petal_width',kde=True,hue='species')
```

```python
# distplot
sns.distplot(x=data['sepal_length'])
```

```python
# pairplot
sns.pairplot(data=data,hue='species')
```
