#### Data Visualization

Through plotting various graphs we can easily understand the relationship in the data


- There are 2 visualization libaries

 1. matplotlib
 2. seaborn

```python
# import all the libraries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
```

```python
data
```

#### Relational plot

1. line
2. scatterplot
3. rel

- the basic requirement of these plots are 2 continuous numerical colmns

```python
plt.figure(figsize=(5,3))
sns.lineplot(x = data['petal_length'],y=data['petal_width'])
```

Insight:

```python
plt.figure(figsize=(5,3))
sns.lineplot(x = data['petal_length'],y=data['petal_width'],ci=None)
# ci = confidence interval
```

```python
plt.figure(figsize=(5,3))
sns.lineplot(x = data['petal_length'],y=data['petal_width'],ci=None, hue=data['species'])
```

```python
# scatter plot
plt.figure(figsize=(5,3))
sns.scatterplot(x = data['petal_length'],y=data['petal_width'])
```

```python
# scatter plot
plt.figure(figsize=(5,3))
sns.scatterplot(x = data['petal_length'],y=data['petal_width'],hue=data['species'])
```

```python
# rel plot
plt.figure(figsize=(5,3))
sns.relplot(x = data['petal_length'],y=data['petal_width'],hue=data['species'])
```

```python
# rel plot
# by defualt is a scatter plot
# convert rel into line plot
plt.figure(figsize=(5,3))
sns.relplot(x = data['petal_length'],y=data['petal_width'],hue=data['species'],kind='line',ci=None)
```

### Distribution graphs

- These are the graphs which are there to analyse the single numerical column

1. histplot / distpolt - there to analyse the single numerical column, and to see the distribution

2. boxplot - will also analyse the numerical col , tell us whther the data has outlier or not

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(5,3))
sns.histplot(x=data['sepal_width'],kde=True)
#kde - kernel density estimator
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(x=data['sepal_width'])
```

### Categorical plot


1 numerical col and 1 categorical

```python
# countplot
# analyse the single categorical data
plt.figure(figsize=(5,3))
sns.countplot(x=data['species'])
```

```python
# analyse 1 cat vs 1 numerical
plt.figure(figsize=(5,3))
sns.barplot(y = data['sepal_width'],x=data['species'])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,3))
sns.boxplot(x = data['sepal_width'],y=data['species'])
```

```python
plt.figure(figsize=(5,3))
sns.violinplot(x = data['sepal_width'],y=data['species'])
```

```python
plt.figure(figsize=(5,3))
sns.stripplot( y = data['sepal_width'],x=data['species'],hue=data['species'])
```

```python
plt.figure(figsize=(5,3))
sns.swarmplot( y = data['sepal_width'],x=data['species'],hue=data['species'])
```

```python
#cat can be converted into strip , swarm , box
#plt.figure(figsize=(5,3))
sns.catplot( y = data['sepal_width'],x=data['species'],hue=data['species'],kind='strip')
```
