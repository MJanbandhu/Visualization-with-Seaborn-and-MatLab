## Exploratory Data Analysis

### Relational Plots

Relational plots are used to understand relationship between two variable

- line plot
- scatter plot
- relplot

```python
# line plot
sns.lineplot(data=df,x='sepal_length',y='sepal_width')
```

```python
# line plot
sns.lineplot(data=df,x='sepal_length',y='sepal_width',ci=None)
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=df,hue='species',ci=None)
```

### Scatter Plot

```python
#scatter plot
sns.scatterplot(x='petal_width',y='petal_length',data=df)
```

```python
sns.scatterplot(x='petal_width',y='petal_length',data=df,hue='species')
```

### Rel plot

```python
#relplot by default is scatter plot
# it can be converted to into the plot giving kind parameter
sns.relplot(data=df,x='petal_length',y='sepal_width')
```

```python
sns.relplot(data=df,x='petal_length',y='sepal_width',kind='line')
```

```python
sns.relplot(data=df,x='petal_length',y='sepal_width',hue='species')
```

## Categorical Plots

It is used to analyze one numerical variable and one categorical variable/columns.

- count plot
- box plot
- bar plot
- violin plot
- strip plot
- swarm plot
- cat plot

### count plot

```python
df.species.value_counts()
```

```python
# It counts categories of the columns and return  count of their occurences.
sns.countplot(x='species',data=df)
# countplot is used to check whether target variable is balanced or not.
```

### Bar Plot

```python
# To plot a bar plot we use categorical column for X axis and numerical column
# for Y axis
# creates a plot taking mean of the categorical column
sns.barplot(x='species',y='sepal_length',data=df)
```

### Box Plot

```python
# It is used to represent the numerical data.
# Boxplot is used to check the outliers in the dataset
sns.boxplot(x='sepal_width',data=df)
```

```python
sns.boxplot(x='species',y='sepal_width',data=df)
```

```python
# violin plot
sns.violinplot(x='sepal_width',data=df)
```

```python
#It is similar to the boxplot except that it provides more advanced visualization.
sns.violinplot(x='species',y='sepal_length',data=df)
```

### Stripplot

```python
#A strip plot can be drawn on its own, but it is also a good complement
#to a box or violin plot in cases where you want to show all observations
#along with some representation of the underlying distribution.
sns.stripplot(x='species',y='sepal_width',data=df,hue='species')
```

### swarmplot

```python
#This gives a better representation of the distribution of values
sns.swarmplot(x='species',y='sepal_width',data=df)
```

### catplot

```python
#using this plot we can plot violin,boxplot,swarmplot
sns.catplot(x='petal_length',data=df)
# by default it will plot scatter plot
```

```python
sns.catplot(x='petal_length',data=df,kind='violin')
```

```python
sns.catplot(x='petal_length',data=df,kind='swarm',hue='species')
```

```python
#task:
#plot strip plot using cat plot
sns.catplot(x='petal_length', data=df, kind='strip', hue='species')
```

## Distribution Plots

### Histogram

```python
# It is used to represent numerical data
sns.histplot(data=df,x='sepal_length')
```

```python
sns.histplot(data=df,x='sepal_length',kde=True)
```

```python
sns.histplot(data=df,x='sepal_length',hue='species')
```

### Distplot

```python
# It is used to do univariate analysis
# It is used to check the distribution of the data
sns.distplot(df['sepal_width'])
```

```python
sns.distplot(df['petal_width'])
```

```python
# Task: check the distribution of petal_length column
sns.distplot(df['petal_length'])
```

### Pairplot

```python
# it represents pairwise relation across the entire dataframe.
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df,hue='species')
```

### Heatmap

```python
# It is used to check the correlation between different featurs
#It is used in feature selection
corr1 = df.corr()
print(corr1)
```

```python
sns.heatmap(corr1,annot=True)
```

```python
df1=df.select_dtypes(include='number').corr()
```

```python
sns.heatmap(df1,annot=True)
```
